```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# RPM ve Torque değerleri (araç motoruna özgü veriler kullanılabilir)
rpm = np.linspace(1000, 6000, 100)
torque = np.linspace(10, 200, 100)

# Meshgrid oluşturma (RPM ve Torque arasında bir yüzey oluşturmak için)
RPM, Torque = np.meshgrid(rpm, torque)

# Verimlilik modeli (Daha karmaşık veya gerçekçi bir model kullanılabilir)
efficiency = (np.sin(RPM / 1000) * np.cos(Torque / 50) * 100) + 50  # Basit bir model, verileri ayarlayabilirsiniz

# 3D Grafik oluşturma
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Yüzey grafiği (Surface Plot) oluşturma
surf = ax.plot_surface(RPM, Torque, efficiency, cmap='viridis', edgecolor='none')

# Renk çubuğu ekleme
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

# Eksen başlıkları
ax.set_xlabel('RPM')
ax.set_ylabel('Torque (Nm)')
ax.set_zlabel('Efficiency (%)')

# Grafik başlığı
plt.title('Motor Verimliliği, RPM ve Tork İlişkisi')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# RPM ve Torque değerleri
rpm = np.linspace(1000, 6000, 100)
torque_levels = [50, 100, 150, 200]  # Sabit tork seviyeleri için analiz

# Verimlilik fonksiyonu (Bu basit model, aracınıza özgü verilerle değiştirilebilir)
def efficiency_model(rpm, torque):
    return (np.sin(rpm / 1000) * np.cos(torque / 50) * 100) + 50

# Verimlilik değerlerini hesaplama ve çizim
plt.figure(figsize=(10, 6))
for torque in torque_levels:
    efficiency = efficiency_model(rpm, torque)
    plt.plot(rpm, efficiency, label=f'Tork: {torque} Nm')

plt.title('Motor Verimliliği - RPM İlişkisi (Farklı Tork Seviyeleri)')
plt.xlabel('RPM')
plt.ylabel('Verimlilik (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Sabit RPM seviyeleri için tork ve verimlilik ilişkisi
torque = np.linspace(10, 200, 100)
rpm_levels = [2000, 3000, 4000, 5000]  # Sabit RPM seviyeleri

plt.figure(figsize=(10, 6))
for rpm_value in rpm_levels:
    efficiency = efficiency_model(rpm_value, torque)
    plt.plot(torque, efficiency, label=f'RPM: {rpm_value} devir')

plt.title('Motor Verimliliği - Tork İlişkisi (Farklı RPM Seviyeleri)')
plt.xlabel('Tork (Nm)')
plt.ylabel('Verimlilik (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Verimlilik matrisini oluşturalım
RPM, Torque = np.meshgrid(rpm, torque)
efficiency = efficiency_model(RPM, Torque)

plt.figure(figsize=(10, 6))
sns.heatmap(efficiency, xticklabels=np.round(rpm, -2), yticklabels=np.round(torque, -1), cmap='viridis')
plt.title('Motor Verimliliği Haritası (RPM ve Tork)')
plt.xlabel('RPM')
plt.ylabel('Tork (Nm)')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Yüzey grafiği (Surface Plot)
surf = ax.plot_surface(RPM, Torque, efficiency, cmap='plasma', edgecolor='none')

# Renk çubuğu ekleme
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

# Eksen başlıkları
ax.set_xlabel('RPM')
ax.set_ylabel('Tork (Nm)')
ax.set_zlabel('Verimlilik (%)')

# Grafik başlığı
plt.title('Motor Verimliliği - RPM ve Tork İlişkisi (3D)')
plt.show()
```
